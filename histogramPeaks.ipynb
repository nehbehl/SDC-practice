```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('warped-example.jpg')/255



def histogram(image):
    bottom_half=img[image.shape[0]//2:,:]
    
    summation_x= np.sum(bottom_half,axis=0)
    print(summation_x.shape[0:2])
    return summation_x

hist=histogram(img)
plt.plot(hist)
```
